```python
# Q no 6-- Create a new column Qtr using numpy.where() or any suitable utility in the imported 
# dataset. The Quarters are based on months and defined as -
#  Jan - Mar : Q1
#  Apr - Jun : Q2
#  Jul - Sep : Q3
#  Oct - Dec : Q4 
```

```python
sales.head()
```

```python
month =sales["Month"]
```

```python
quarter = []
for x in month :
    if x in ["Jan","Feb","Mar"]:
        quarter.append("Q1")
    elif x in ["Apr","May","Jun"]:
        quarter.append("Q2")
    elif x in ["Jul","Aug","Sep"]:
        quarter.append("Q3")
    else:
        quarter.append("Q4")
```

```python
quarter
```

```python
sales["Qtr"]= pd.Series(quarter)
```

```python
sales.head()
```

```python
## Q no-- 7  Compare Qtr wise sales in 2015 and 2016 in a bar plot
```

```python
qtr_sales = sales.groupby("Qtr")["Sales2015","Sales2016"].sum()
```

```python
qtr_sales
```

```python
qtr_sales.plot(kind="bar")
plt.ylabel("Sales")
plt.title("Quarter wise sales in 2015 and 2016")
plt.show()
```
